```r
geno <-read.table("Hetero_realigned_cov10_filtered3.raw", row.names=1, header=T)
pheno <- read.csv("2017heteroPheno.csv", header=T)
train_id<-read.csv("idtrain")
test_id<-read.csv("idtest")
attach(pheno)
x <- as.matrix(geno)-1             #Convert geno file to matrix
```

```r
library(sommer)
```

```r
row.names(geno)=1:240
A <- A.mat(geno)
data <- data.frame(bcw=bcw,length=length,id=factor(1:240))
```

```r
ans.A <- mmer2(cbind(length,bcw)~1,
               random=~us(trait):g(id),
               rcov=~us(trait):units,
               G=list(id=A), 
               data=data,
               silent = TRUE)
```

```r
summary(ans.A)
```

```r
gvc<- ans.A$var.comp$'g(id)'
sd.gvc <- as.matrix(sqrt(diag(gvc)))
prod.sd <- sd.gvc %*% t(sd.gvc)
(gen.cor <- gvc/prod.sd)
```
