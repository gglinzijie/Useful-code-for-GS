```python
!activate py36
import matplotlib.pyplot as plt
import numpy as np
from pystrich.code39 import Code39Encoder
```

```python
begin=6
end=18
length=end-begin
space=0.1
d={}
m=[]
```

```python
for x in np.arange(begin, end, space):
    value=round(x,1)
    encoder = Code39Encoder(f"{value}")
    encoder.save( f"code39_{value}.png" )
for x in np.arange(begin, end, space):
    value=round(x,1)
    d["im{0}".format(value)]=plt.imread(f"code39_{value}.png")
```

```python
for i,j in d.items():
    m.append(i)
m1=np.array(m)
input=np.transpose(m1.reshape(length,10)).flatten()
```

```python
fig, axs = plt.subplots(nrows=10, ncols=length, figsize=(120, 60),sharex=True,sharey=True,
                        subplot_kw={'xticks': [], 'yticks': []})

fig.subplots_adjust(left=0.1, right=0.9, hspace=0.3, wspace=0.1)

for ax, key in zip(axs.flat, input):
    ax.imshow(d[key], cmap=plt.cm.bone)

plt.tight_layout()
plt.show()
```

```python
fig.savefig('pict.png')
```
